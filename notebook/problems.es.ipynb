## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import numpy as np
import pandas as pd
```

```python
n_experimentos = 1000
eventos = 0

for l in range(n_experimentos):

    dado1 = np.random.randint(1,7)
    dado2 = np.random.randint(1,7)

    suma_dados = dado1 + dado2

    if (suma_dados > 7) or ((suma_dados % 2) == 0):
        eventos+=1

prob = eventos/n_experimentos
print(prob)
```

```python
def experimentos_proba(n_experimentos):

    eventos = 0

    for l in range(n_experimentos):

        dado1 = np.random.randint(1,7)
        dado2 = np.random.randint(1,7)

        suma_dados = dado1 + dado2

        if (suma_dados > 7) or ((suma_dados % 2) == 0):
            eventos+=1
    prob = eventos/n_experimentos
    return prob

experimentos_proba(1000)
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
import random

ball_box = {}

# Crea la cajas con las bolas
for i in range(60): 
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)


 


```

```python
# Extracción de 5 bolas

def extracción5(experimentos):
   condicion1 = 0
   condicion2 = 0
   
   for e in range(experimentos):
      cinco_colores = []
      for b in range(5):
         elegido = random.choice(ball_box)
         cinco_colores.append(elegido)

      blancas = cinco_colores.count("White")  
      rojas = cinco_colores.count("Red")
      verdes = cinco_colores.count("Green")
      if blancas == 3 and rojas == 2:
         condicion1 = condicion1+1
      elif (blancas == 5) or (rojas == 5) or (verdes == 5):
         condicion2 = condicion2+1

   return condicion1/(experimentos), condicion2/(experimentos)

print(f"La probabilidad de extraer 3 blancas y 2 rojases es de {extracción5(10000)[0]}")
print(f"La probabilidad de extraer las 5 bolas del mismo color es de {extracción5(10000)[1]}")
```

```python
extracción5(10000)
```
